```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

Câu 01 :

```python
data = [
    ['Alice', 25, 50000],
    ['Bob', 30, 60000], 
    ['Charlie', 35, 70000],
    ['David', 28, 55000],
    ['Eva', 22, 52000],
    ['Frank', 45, 80000],
    ['Grace', 34, 72000],
    ['Hannah', 31, 68000],
    ['Ivan', 27, 61000],
    ['Jack', 29, 59000],
    ['Kelly', 33, 63000], 
    ['Liam', 40, 77000],
    ['Mona', 26, 53000],
    ['Nina', 32, 66000],
    ['Oscar', 36, 75000]
]
```

```python
df = pd.DataFrame(data, columns=['NAME', 'AGE', 'SALARY'])
df
```

Câu 02 :

```python
df.info()
```

Câu 03 :

```python
cau03 = df[(df["AGE"] > 28)]
cau03
```

Câu 04 :

```python
cau04 = df['SALARY'].agg('mean')
cau04
```

Câu 05 :

```python
cau05 = df.groupby('AGE')['SALARY'].agg('sum')
cau05
```

Câu 06 :

```python
cau06 = df.sort_values(by='SALARY', ascending=False)
cau06
```

Câu 07 :

```python
import matplotlib.pyplot as plt

df['AGE'].plot(kind='bar')
plt.title('Age Distribution')
plt.xlabel('NAME')
plt.ylabel('AGE')
plt.show()
```

Câu 08 :

```python
df['SALARY'].plot(kind='line')
plt.title('Salary Distribution')
plt.xlabel('NAME')
plt.ylabel('SALARY')
plt.show()
```

Câu 09 :

```python
cau09 = df['AGE'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(cau09, labels=cau09.index, autopct='%1.1f%%')
plt.title('Age Distribution')
plt.axis('equal')  
plt.show()
```

Câu 10 :

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['AGE'], df['SALARY'])
plt.title('AGE vs SALARY')
plt.xlabel('AGE')
plt.ylabel('SALARY')
plt.show()
```
